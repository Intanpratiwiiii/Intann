```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

#membaca gambar
image = cv2.imread ('Strawberry.png',cv2.IMREAD_GRAYSCALE)

#dapatkan resolusi gambar dalam piksel
width, height = image.shape
print (f"Resolusi gambar : {width}x{height} piksel")

#menampilkan gambar
plt.figure(figsize=(5,5))
plt.imshow(image, cmap='gray')
plt.title('Gambar Grayscale')
plt.show()
```

```python
#image data representation
print('display the image matrix contains',len(image[:2]),'pixels')
print(image[:2]) #[start:stop]-- select the row data
#print(image[:,:2])
```

```python
#save image data to txt files
pixel_data = np.array(image)

#saving the 2D array in a text file
np.savetxt('image_data_percobaan.txt',pixel_data,fmt='%d')
```

```python
#jika ingin melihat sinyal dari seluruh gambar (2D sinyal)
signal_2d = image.flatten() #mengubah gambar menjadi vektor 1D

plt.figure(figsize=(12,6))
plt.plot(signal_2d)
plt.title('sinyal dari seluruh gambar (datar)')
plt.xlabel('indeks piksel')
plt.ylabel('intensitas')
plt.show()
```

```python
#3D plotting
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

#create 3d plot
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection= '3d')
ax.grid()

#image data
x = np.arange(image.shape[1])
y = np.arange(image.shape[0])
x,y = np.meshgrid(x,y)

#plot the image data
ax.plot_surface(x,y,image,cmap='gray')

#set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('intensitas')
ax.set_title('gambar grayscale')
plt.show()
```

```python
#downsampling process
factors = 3**np.arange(1,4)

#create a subplot for each sample
figure,axis = plt.subplots(1,len(factors),figsize=(15,5))

#loop over each sampling factor
for factor,ax in zip(factors,axis):
  new_size = (image.shape[1] // factor,image.shape[0] // factor)
  downsampled_image = cv2.resize(image,new_size)
  ax.imshow (downsampled_image,cmap='gray')
  ax.set_title(f'gambar hasil downsampled (factor = {factor})')
plt.show()
```

```python
#upsampling process
factors = 3**np.arange(1,4)

#create a subplot for each sample
figure,axis = plt.subplots(1,len(factors),figsize=(15,5))

#loop over each sampling factor
for factor,ax in zip(factors,axis):
  new_size = (image.shape[1] * factor,image.shape[0] * factor)
  upsampled_image = cv2.resize(image,new_size)
  ax.imshow (upsampled_image,cmap='gray')
  ax.set_title(f'gambar hasil upsampled (factor = {factor})')
  print (f'image size for factor {factor}', upsampled_image.shape)
plt.show()
```

Misalnya, di bagian bawah gambar yang aslinya memiliki nilai piksel 0 dan 1, setelah shrinking terdapat piksel dengan nilai antara (seperti 2 dan 3), yang disebabkan oleh interpolasi antar piksel.

Kesimpulan:
  - Downsampling dan shrinking menyebabkan pengurangan ukuran gambar yang signifikan, yang terlihat dari perubahan ukuran gambar asli (224x225) menjadi lebih kecil (100x100).

  - Dalam proses shrinking, interpolasi bicubic mencoba menghasilkan nilai piksel baru dengan mempertimbangkan piksel di sekitarnya, sehingga menghasilkan gradasi warna yang lebih halus daripada gambar asli yang terlihat lebih tajam pada transisi nilai piksel. Ini menyebabkan adanya perbedaan nilai piksel setelah shrinking.

  - Proses ini juga menunjukkan bagaimana detail pada gambar dapat berkurang seiring dengan penurunan resolusi, meskipun interpolasi digunakan untuk menjaga transisi antar piksel.
