# Data Visualization

### Univariate Analysis

```python
plt.figure(figsize=(10,6))
sns.distplot(df.body_mass_g,color='r')
plt.title('body_mass_g Distribution',size=18)
plt.xlabel('body_mass_g',size=14)
plt.ylabel('Density',size=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.distplot(df['year'])
plt.title('year Distribution',size=18)
plt.xlabel('year',size=14)
plt.ylabel('Count',size=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df.bill_length_mm,color='y')
plt.title('bill_length_mm Distribution',size=18)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='sex',data=df)
plt.title('Total Number of Male and Female',size=18)
plt.xlabel('Sex',size=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='id',data=df)
plt.title('id Distribution',size=18)
plt.xlabel('id',size=14)
plt.ylabel('Count',size=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='species',data=df)
plt.title('species Distribution',size=18)
plt.xlabel('species',size=14)
plt.ylabel('Count',size=14)
plt.show()
```

```python
df.species.value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='island',data=df)
plt.title('island Distribution',size=18)
plt.xlabel('island',size=14)
plt.ylabel('Count',size=14)
plt.show()
```

### Bivariate Analysis

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='year',y='body_mass_g',data=df)
plt.title('year vs body_mass_g Distribution',size=18)
plt.xlabel('year',size=14)
plt.ylabel('body_mass_g Distribution',size=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='species',y='body_mass_g',data=df)
plt.title('species vs body_mass_g',size=18)
```

```python
sns.pairplot(df,
             markers="+",
             diag_kind="kde",
             kind='reg',
             plot_kws={'line_kws':{'color':'#aec6cf'},
                       'scatter_kws':{'alpha':0.7,
                                      'color': 'red'}},
             corner=True);
```

### Multivariate Analysis

Correlation

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=np.number).corr(),annot=True,square=True)
plt.title('Correlations Between Variables',size=18);
plt.xticks(size=13)
plt.yticks(size=13)
plt.show()
```
