```python
import cv2
import numpy as np
from scipy.stats.kde import gaussian_kde
import matplotlib.pyplot as plt
# Import the cv2_imshow function from google.colab.patches
from google.colab.patches import cv2_imshow

# original image
f = cv2.imread('lenna.png', 0)
f = f/255

# Use cv2_imshow instead of cv2.imshow
cv2_imshow(f) # Display the image using cv2_imshow
#cv2.waitKey(0) # These lines are not needed with cv2_imshow
#cv2.destroyAllWindows()

# create gaussian noise
x, y = f.shape
mean = 0
var = 0.01
sigma = np.sqrt(var)
n = np.random.normal(loc=mean,
                     scale=sigma,
                     size=(x,y))

# Use cv2_imshow instead of cv2.imshow
cv2_imshow(n) # Display the image using cv2_imshow
#cv2.waitKey(0) # These lines are not needed with cv2_imshow
#cv2.destroyAllWindows()

# display the probability density function (pdf)
kde = gaussian_kde(n.reshape(int(x*y)))
dist_space = np.linspace(np.min(n), np.max(n), 100)
plt.plot(dist_space, kde(dist_space))
plt.xlabel('Noise pixel value'); plt.ylabel('Frequency')
plt.show()

# add a gaussian noise
g = f + n

# Use cv2_imshow instead of cv2.imshow
cv2_imshow(g) # Display the image using cv2_imshow
#cv2.waitKey(0) # These lines are not needed with cv2_imshow
#cv2.destroyAllWindows()

# display all
cv2_imshow(f) # Display the image using cv2_imshow
cv2_imshow(n) # Display the image using cv2_imshow
cv2_imshow(g) # Display the image using cv2_imshow

#cv2.waitKey(0) # These lines are not needed with cv2_imshow
#cv2.destroyAllWindows()
```
